```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype = float)
farenheit = np.array([-40, 14, 32, 46.4, 59, 71.6, 100.4], dtype = float)
```

```python
#dise√±ando la red neuronal simple
#neuronas de la capa, 
capa = tf.keras.layers.Dense(units = 1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Comenzando entrenamiento")
historial = modelo.fit(celsius,farenheit,epochs=1000, verbose=False)
print("Modelo entrenado")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud perdida")
plt.plot(historial.history["loss"])
```

```python
print("Haciendo una prueba")
resultado = modelo.predict([100.0])
print("El resultado es: "+str(resultado)+ "fahrenheit")
```
